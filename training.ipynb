```python
from scipy.signal import butter, filtfilt

def bandpass_filter(segment, fs=360, lowcut=0.5, highcut=45, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, segment, axis=0)
```

```python
record_list = [100, 101, 102,103,104,105,106,107,108,109,111,112,113,114,115,116,117,118,119,121,122,123,124,200,201,202,203,205,207,208,209,210,212,213,214,215,217,219,220,221,222,223,228,230,231,232,233,234]
segments = []
labels = []

window_size = 180  # 0.5 seconds if records are at 360 Hz (360 * 0.5)
channels = 2       # MLII and V5

for rec in record_list:
    
    db_path = r"C:\Me\College\4th\Ai_in_healthcare\Ai_in_healthcare_project\database"
    record_name = db_path + f"\\{rec}"
    record = wfdb.rdrecord(record_name)
    annotation = wfdb.rdann(record_name, 'atr')
    signal = record.p_signal
    
    for idx, sample_idx in enumerate(annotation.sample):
        # Only extract windows where enough data exists before/after the point
        if sample_idx - window_size//2 >= 0 and sample_idx + window_size//2 < signal.shape[0]:
            seg = signal[sample_idx - window_size//2 : sample_idx + window_size//2]
            if seg.shape[0] == window_size:  # Ensure segment length
                filtered_seg = bandpass_filter(seg, fs=360, lowcut=0.5, highcut=45, order=4)
                segments.append(filtered_seg)
                labels.append(annotation.symbol[idx])
```

```python
import numpy as np
# --- Mapping Definition ---
AAMI_MAPPING = {
    # Class N (Normal Beats -> 0)
    'N': 0, 'L': 0, 'R': 0, 'e': 0, 'j': 0, '/': 0,  # '/' (Paced) often grouped with N
    
    # Class S (Supraventricular Ectopic Beats -> 1)
    'A': 1, 'a': 1, 'J': 1, 'S': 1,
    
    # Class V (Ventricular Ectopic Beats -> 2)
    'V': 2, 'E': 2,
    
    # Class F (Fusion Beats -> 3)
    'F': 3, 'f': 3,
    
    # Class Q (Unclassifiable/Other Beats -> 4)
    'Q': 4,
    
    # Non-Beat/Noise Annotations (To be filtered out -> -1)
    '!': -1, '"': -1, '+': -1, '[': -1, ']': -1,
    '|': -1, 'x': -1, '~': -1
}
# At this point:
# segments is a list of NumPy arrays (shape: (180, 2))
# labels is a list of characters (e.g., ['N', 'V', 'N', '!', 'A', ...])

print(f"Total raw labels before consolidation: {len(labels)}")
# print(f"Example raw labels: {labels[:10]}")

# 1. Convert the list of raw labels to a NumPy array for vectorized mapping
raw_labels_array = np.array(labels, dtype=object) # Use object dtype for characters

# 2. Apply the mapping to get consolidated numerical classes
# np.vectorize makes the dictionary lookup work across the entire array
get_aami_class = np.vectorize(lambda x: AAMI_MAPPING.get(x, -1)) # Default to -1 if symbol is missing
consolidated_labels = get_aami_class(raw_labels_array)

# 3. Create a boolean mask to filter out "useless" labels (where class is -1)
filter_mask = consolidated_labels != -1

# 4. Apply the mask to BOTH the segments and the labels
# This ensures your segments and labels lists remain perfectly synchronized!
final_segments = np.array(segments)[filter_mask]
final_labels = consolidated_labels[filter_mask]

print(f"Total segments and labels remaining after filtering: {len(final_segments)}")
print(f"Shape of final segments array: {final_segments.shape}")
print(f"Unique classes remaining in final_labels: {np.unique(final_labels)}")
```

```python
import numpy as np
import pandas as pd 
from collections import Counter

# --- Function to Analyze and Print Class Distribution ---

def analyze_class_distribution_simple(y_train):
    """
    Calculates class counts, percentages, and prints the distribution of y_train
    in a simple, easy-to-read table.
    
    Args:
        y_train (np.array): The 1D NumPy array of integer labels (0, 1, 2, 3, 4).
    """
    if y_train.ndim != 1:
        print("Error: y_train must be a 1-dimensional array of integer labels.")
        return

    # 1. Map integer classes back to their descriptive names (AAMI Standard)
    AAMI_CLASS_NAMES = {
        0: 'N (Normal)',
        1: 'S (Supraventricular)',
        2: 'V (Ventricular)',
        3: 'F (Fusion)',
        4: 'Q (Unclassifiable)'
    }

    # 2. Get counts and percentages using Pandas for efficiency
    label_series = pd.Series(y_train).astype(int)
    # Value counts automatically sorts by count descending, we sort by index (class number)
    counts = label_series.value_counts().sort_index()
    total_samples = len(y_train)
    percentages = (counts / total_samples) * 100

    print("\n" + "=" * 50)
    print("        AAMI ECG Class Distribution in y_train")
    print("=" * 50)
    
    # Print the table header
    print(f"{'Class (AAMI)':<25} | {'Count':<10} | {'Percentage':<10}")
    print("-" * 50)
    
    # Print data rows
    for class_int in counts.index:
        name = AAMI_CLASS_NAMES.get(class_int, f'Unknown ({class_int})')
        count = counts[class_int]
        percent = percentages[class_int]
        # Highlight the Normal class (0) as it dominates
        prefix = "-> " if class_int == 0 else "   "
        print(f"{prefix}{name:<22} | {count:<10} | {percent:.2f}%")
        
    print("-" * 50)
    print(f"{'TOTAL SAMPLES':<25} | {total_samples:<10} | {'100.00%':<10}")
    print("=" * 50)


# --- Example Usage (Run this after your data split) ---
# NOTE: You MUST replace 'y_train_simulated' with your actual y_train NumPy array
# For demonstration, we use the simulated data:
# y_train = np.concatenate([
#     np.zeros(80000),  # Class 0 (N)
#     np.ones(3000),    # Class 1 (S)
#     np.full(4000, 2), # Class 2 (V)
#     np.full(500, 3),  # Class 3 (F)
#     np.full(74, 4)    # Class 4 (Q)
# ]).astype(int)
# np.random.shuffle(y_train)
analyze_class_distribution_simple(y_train)
```

```python

import tensorflow as tf

from imblearn.over_sampling import SMOTE

def resample_data(X_train, y_train):
    """
    Performs SMOTE oversampling to balance the training data.
    This creates synthetic samples for minority classes to match the majority.
    """
    
    # 1. Flatten the data for SMOTE
    # SMOTE expects 2D data: (n_samples, n_features)
    n_samples, n_timesteps, n_features = X_train.shape
    X_flattened = X_train.reshape(n_samples, n_timesteps * n_features)
    
    print("-" * 50)
    print(f"SMOTE: Flattening data from {X_train.shape} to {X_flattened.shape}")
    
    # 2. Apply SMOTE
    # The 'not majority' strategy balances all minority classes up to the majority count.
    # We remove n_jobs=-1 as it caused an error before.
    smote = SMOTE(sampling_strategy='not majority', random_state=42)
    
    print("SMOTE: Starting oversampling (this may take a few minutes)...")
    X_resampled_flattened, y_resampled = smote.fit_resample(X_flattened, y_train)
    
    # 3. Reshape X_resampled back to 3D
    # New shape: [N_resampled_samples, 180, 2]
    X_resampled = X_resampled_flattened.reshape(-1, n_timesteps, n_features)

    print(f"SMOTE: Resampling complete.")
    print(f"Original samples: {n_samples} -> Resampled samples: {X_resampled.shape[0]}")
    print("-" * 50)
    
    # 4. Analyze the new (balanced) distribution
    analyze_class_distribution_simple(y_resampled)
        
    # 5. One-Hot Encode the resampled labels
    num_classes = np.unique(y_resampled).size
    y_resampled_encoded = tf.keras.utils.to_categorical(y_resampled, num_classes=num_classes)
    
    print(f"Final X_resampled shape: {X_resampled.shape}")
    print(f"Final y_resampled_encoded shape: {y_resampled_encoded.shape}")
    print("-" * 50)
    
    return X_resampled, y_resampled_encoded
```

```python
import tensorflow as tf
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dropout, LSTM, Dense, Flatten
# from tensorflow.keras.optimizers import Adam
import numpy as np

# Define the deep, high-capacity CNN-LSTM model
def build_optimal_cnn_lstm_model(input_shape, num_classes):
    """
    Builds an optimized CNN-LSTM model with strong regularization 
    (Dropout, Batch-Norm, and L2) to combat overfitting.
    """
    
    print(f"\nBuilding optimal model with input shape: {input_shape} and {num_classes} classes.")
    
    l2_reg = tf.keras.regularizers.l2(0.001) # Define L2 regularizer

    inputs = tf.keras.layers.Input(shape=input_shape)
    
    # --- CNN BLOCK 1 ---
    x = tf.keras.layers.Conv1D(filters=64, kernel_size=10, activation='relu', padding='same', kernel_regularizer=l2_reg)(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=3, strides=2, padding='same')(x)
    x = tf.keras.layers.Dropout(0.2)(x) 

    # --- CNN BLOCK 2 ---
    x = tf.keras.layers.Conv1D(filters=128, kernel_size=7, activation='relu', padding='same', kernel_regularizer=l2_reg)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=3, strides=2, padding='same')(x)
    x = tf.keras.layers.Dropout(0.2)(x) 

    # --- CNN BLOCK 3 ---
    x = tf.keras.layers.Conv1D(filters=256, kernel_size=5, activation='relu', padding='same', kernel_regularizer=l2_reg)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=3, strides=2, padding='same')(x)
    x = tf.keras.layers.Dropout(0.3)(x) 

    # --- LSTM BLOCK: Sequence Learning ---
    x = tf.keras.layers.LSTM(128, return_sequences=True, kernel_regularizer=l2_reg)(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    
    x = tf.keras.layers.LSTM(128, kernel_regularizer=l2_reg)(x)
    x = tf.keras.layers.Dropout(0.3)(x)

    # --- CLASSIFICATION HEAD: Final Prediction ---
    x = tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=l2_reg)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.5)(x) 

    outputs = tf.keras.layers.Dense(num_classes, activation='softmax')(x)
    
    model = tf.keras.models.Model(inputs=inputs, outputs=outputs)

    # --- COMPILE WITH LOWER LEARNING RATE ---
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001) 
    
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

```python
import tensorflow as tf
import numpy as np # (Assuming y_test is already a NumPy array)


X_train_final, y_train_final = resample_data(X_train,y_train)
num_classes = 5 
y_test_encoded = tf.keras.utils.to_categorical(y_test, num_classes)
print(y_train_final.shape)
print(X_train_final.shape)
```

```python
input_shape = X_train_final.shape[1:]  # e.g., (180, 2)
num_classes = y_train_final.shape[1] # e.g., 5


model = build_optimal_cnn_lstm_model(input_shape, num_classes)
early_stopping = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', # Monitors the validation loss
        patience=15,        # Number of epochs with no improvement after which training will be stopped
        restore_best_weights=True # Restores model weights from the epoch with the best value of the monitored quantity
    )
Heart_disease_detector = model.fit(
    X_train_final,  # <-- Use the original, imbalanced data
    y_train_final, # <-- Use the original, encoded labels
    epochs=50,
    batch_size=256,
    validation_data=(X_test, y_test_encoded),
    callbacks=[early_stopping],
)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
def evaluate_model(model, X_test, y_test_encoded):
    """
    Performs a full evaluation on the test set, printing a
    classification report and plotting a confusion matrix.
    """
    print("\nEvaluating model on the test set...")
    
    # AAMI Class Names (for reports)
    AAMI_CLASS_NAMES = ['N (Normal)', 'S (Supraventricular)', 'V (Ventricular)', 'F (Fusion)', 'Q (Unclassifiable)']
    
    # A. Get final loss and accuracy
    test_loss, test_accuracy = model.evaluate(X_test, y_test_encoded, verbose=0)
    print(f"\nTest Set Loss:     {test_loss:.4f}")
    print(f"Test Set Accuracy: {test_accuracy*100:.2f}%")
    
    # B. Generate predictions (probabilities)
    y_pred_proba = model.predict(X_test)
    
    # C. Convert probabilities to hard class labels (e.g., [0.1, 0.8] -> 1)
    y_pred = np.argmax(y_pred_proba, axis=1)
    # Also convert one-hot encoded y_test back to single labels
    y_true = np.argmax(y_test_encoded, axis=1)

    # D. Print Classification Report (Precision, Recall, F1-Score)
    print("\n" + "="*50)
    print("      CLASSIFICATION REPORT (Test Set)")
    print("="*50)
    print(classification_report(y_true, y_pred, target_names=AAMI_CLASS_NAMES))
    
    # E. Generate and Plot Confusion Matrix
    print("\n" + "="*50)
    print("      CONFUSION MATRIX (Test Set)")
    print("="*50)
    
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=AAMI_CLASS_NAMES, 
                yticklabels=AAMI_CLASS_NAMES)
    plt.title('Confusion Matrix (True Labels vs. Predicted Labels)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```
